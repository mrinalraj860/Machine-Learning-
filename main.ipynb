```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
from sklearn.metrics import accuracy_score
from nltk.stem.porter import PorterStemmer
```

```python
port=PorterStemmer()
```

```python
def stemming(content):
    review = re.sub('[^a-zA-Z]',' ',str(content))
    review = review.lower()
    review = review.split()
    review = [port.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    return review
test['content']=test['content'].apply(stemming)
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(num_words=5000,oov_token="<OOV>")
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
mean=[]
for i in range(len(x_test)):
    mean.append(len(x_test[i]))
```

```python
padded = pad_sequences(x_test, maxlen=50, padding="pre")
print(padded)
```

```python
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Embedding(5000,40,input_length=50))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
x_test=np.array(padded)
y_test=np.array(y_test)
```

```python
model.fit(x_test,y_test,epochs=20,batch_size=64)
```
