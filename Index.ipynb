```python
#from itertools import Predicate
from PIL import Image
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import mnist
#from itertools import Predicate



```

```python
x_train=mnist.train_images()
y_train=mnist.train_labels()


x_test=mnist.test_images()
y_test=mnist.test_labels()

print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)

#as shape is 60000 ,28,28 we need to resize it as 28*28 cause its data of 28pizel so tro convert it to array we need the list of all thr
#pixels


x_train=x_train.reshape(-1,28*28)
x_test=x_test.reshape(-1,28*28)

#as data is from value 0 to 256 we need to have our data between 0 to 1 so that the model can work like classification

x_train=(x_train/256)
x_test=(x_test/256)
'''activation{‘identity’, ‘logistic’, ‘tanh’, ‘relu’}, default=’relu’
Activation function for the hidden layer.



‘identity’, no-op activation, useful to implement linear bottleneck, returns f(x) = x

‘logistic’, the logistic sigmoid function, returns f(x) = 1 / (1 + exp(-x)).

‘tanh’, the hyperbolic tan function, returns f(x) = tanh(x).

‘relu’, the rectified linear unit function, returns f(x) = max(0, x)

solver{‘lbfgs’, ‘sgd’, ‘adam’}, default=’adam’
The solver for weight optimization.

‘lbfgs’ is an optimizer in the family of quasi-Newton methods.

‘sgd’ refers to stochastic gradient descent.

‘adam’ refers to a stochastic gradient-based optimizer proposed by Kingma, Diederik, and Jimmy Ba

'''
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
model=MLPClassifier(solver='adam',activation='relu',hidden_layer_sizes=(64,64))
model.fit(x_train,y_train) #supervised

pric=model.predict(x_test)

accuracy=confusion_matrix(y_test,pric)

def find_acc_from_matrix(acc):
    elem=acc.trace()
    all=acc.sum()
    return elem/all

final_acc=find_acc_from_matrix(accuracy)

print(final_acc)
```
