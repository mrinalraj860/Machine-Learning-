```python
file=pd.read_csv("Student_Marks.csv")
array=np.array(file[1:])
```

```python
x=np.array(array[:,:2])
#print(x)
#print(x.shape)
y=np.array(array[:,2:3])
#print(y)
#print(y.shape)
```

```python
model=sklearn.linear_model.LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.24)
#x_final.reshape(-1,2)
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
print(pred,"  ",y_test)
```

```python
print("R^2 value:" , model.score(x,y))
```
